```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Community_Violence_Intervention_Data.csv")
```

```python
df.shape
```

```python
# Convert the Month_Year column into a proper datetime format (YYYY-MM)
df["Month_Year"] = pd.to_datetime(df["Month_Year"])
```

```python
df
```

```python
#Extract month name and year into separate columns.
df['Month'] = df['Month_Year'].dt.month
df['Year'] = df['Month_Year'].dt.year
```

```python
df
```

```python
# Check for duplicate months per organization and resolve inconsistencies.
duplicate_values = df['Month'].duplicated()
print(duplicate_values)
```

```python
df = df.drop_duplicates(subset=['Month'], keep='first')
```

```python
df
```

```python
#Standardize values in the Organization column (remove trailing spaces, ensure consistent capitalization).

df.loc['Organization'] = df['Organization'].str.strip().str.title()
```

```python
df
```

```python
df = df.copy()  # ensure we're not working on a slice

def clean_split_address(address):
    try:
        # Split from the rightmost comma to separate street/city and state/zip
        street_city, state_zip = address.rsplit(',', 1)
        
        # Clean whitespace
        street_city = street_city.strip()
        state_zip = state_zip.strip()
        
        # State and ZIP
        state_zip_split = state_zip.split(' ', 1)
        state = state_zip_split[0].upper()
        zip_code = state_zip_split[1] if len(state_zip_split) > 1 else None
        
        # City is the last word in street_city
        street_parts = street_city.rsplit(' ', 1)
        street = street_parts[0].title()
        city = street_parts[1].title() if len(street_parts) > 1 else None
        
        return pd.Series({'Street': street, 'City': city, 'State': state, 'Zip': zip_code})
    
    except Exception:
        # Fallback for malformed addresses
        return pd.Series({'Street': None, 'City': None, 'State': None, 'Zip': None})

# Apply to the DataFrame
df[['Street','City','State','Zip']] = df['Address'].apply(clean_split_address)
```

```python
df
```

```python
# 1. Truncate 'Zip+4' and remove non-numeric characters
df['ZipCode_Clean'] = df['ZipCode'].astype(str).str.replace(r'[^\d-]', '', regex=True).str.split('-').str[0].str.zfill(5)

# 2. Validate: Keep only 5-digit codes, set others to NaN
df['ZipCode_Clean'] = df['ZipCode_Clean'].apply(
lambda x: x if pd.notna(x) and len(str(x)) == 5 and str(x).isdigit() else None
 )

```

```python
df.shape
```

```python
# Identify potential integer columns by checking for columns with names suggesting counts or numbers
# For demonstration, let's assume columns containing 'Count', 'Num', 'Mediation', 'Intervention' 
# might be integer columns that need conversion

int_cols = [col for col in df.columns if any(keyword in col.lower() for keyword in ['count', 'num', 'mediation', 'intervention'])]

# Convert these columns to numeric, coercing errors to NaN, then convert to integers (nullable Int64 dtype)
for col in int_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')

df[int_cols].head()
```

```python
# Replace missing or invalid numeric entries in the identified integer columns with 0
df[int_cols] = df[int_cols].fillna(0)
df[int_cols].head()
```

```python
# Detect and remove negative numbers in the identified integer/count columns by replacing them with 0
for col in int_cols:
    df.loc[df[col] < 0, col] = 0

df[int_cols].head()
```

```python
import matplotlib.pyplot as plt

# Example: Check for outliers in the integer/count columns using boxplots
for col in int_cols:
    plt.figure(figsize=(6, 3))
    plt.boxplot(df[col], vert=False)
    plt.title(f'Boxplot for {col}')
    plt.xlabel(col)
    plt.show()

# Additionally, get statistical summary to identify potential outliers
df[int_cols].describe()
```

```python
# Drop the FID column if it is just an index and not meaningful.
df.drop("FID", axis=1)
```
